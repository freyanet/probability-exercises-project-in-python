## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
# Se importan los modulos necesarios
import numpy as np
import random

#Creamos una funcion que tenga como parametro la cantidad de simulaciones que se haran del experimento en este caso 1000
def probabilidad_dados(simulaciones = 1000):
     #Se crea una variable int 0 para luego ir sumando cada que se cumple la condicion que se busca
    suma = 0

    #Se realiza un loop for con el rango de la cantidad de simulaciones
    for i in range(simulaciones): 

        #Creamos los dados para que generen numeros random del 1 al 6 cada vez que se itere
        dado_1 = random.randint(1,6) 
        dado_2 = random.randint(1,6) 

        #Realizamos la suma de los dados en cada iteracion
        prob = dado_1 + dado_2 

        #Creamos una condicional segun que se nos pide, en este caso, numero par o mayor que 7
        if (prob %2 == 0) or (prob > 7): 

            #Agregamos un valor a la variable inicial para luego calcular la probabilidad
            suma += 1 

    #Devolvemos el resultado dividido entre lass simulaciones
    return suma / simulaciones 



print(f"La probabilidad de que el resultado sea par o mayor que 7 es de: {np.round(probabilidad_dados() * 100, 2)}%")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
import random
import numpy as np

ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)

# Creamos un bucle de la cantidad de veces que se ejecutara el experimento
def balls_prob(simulaciones = 1000):

    # Declaramos 2 variables que seran las veces que se cumplan las condiciones solicitadas
    count1 = 0 # 3 blancas y 2 rojas
    count2 = 0 # Todas del mismo color

    for i in range(simulaciones):

        muestra = []

        for j in range(5):
            color_id = np.random.randint(0,59)
            muestra.append(ball_box[color_id])
        
        colors = np.array(muestra)

        whites = sum(colors == 'White')
        reds = sum(colors == 'Red')
        greens = sum(colors == 'Green')

        if (whites == 3) and (reds == 2):
            count1 += 1

        if (whites == 5) or (reds == 5) or (greens == 5):
            count2 += 1
    
    return count1/simulaciones,count2/simulaciones


probabilidades = balls_prob(1000)


print(f"La probabilidad de que salgan 3 blancas y 2 rojas es de: {np.round(probabilidades[0] * 100,2)}%")
print(f"La probabilidad de que salgan todas del mismo color es de: {np.round(probabilidades[1] * 100,2)}%")


```
